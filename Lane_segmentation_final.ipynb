```python
import yaml
from ultralytics import YOLO
from yolov5.val import run as yolov5_val

# --------------------------------------------
# 📁 CONFIGURATION
# --------------------------------------------
DATA_YAML = '/kaggle/working/data_model.yaml'

YOLOV5_MODEL = '/kaggle/working/yolov5/runs/train/yolov5_lane_seg/weights/best.pt'
YOLOV8S_MODEL = '/kaggle/working/runs/segment/yolov8_lane_seg/weights/best.pt'
YOLOV11S_MODEL = '/kaggle/working/runs/segment/yolov11_lane_seg/weights/best.pt'

IMG_SIZE = 640
BATCH_SIZE = 8
DEVICE = '0'

# --------------------------------------------
# 🔍 VALIDATE YOLOv8s-seg (Official)
# --------------------------------------------
print("\n🔎 Validating YOLOv8s-seg (Official)...")
model_v8s = YOLO(YOLOV8S_MODEL)
results_v8s = model_v8s.val(data=DATA_YAML, imgsz=IMG_SIZE, batch=BATCH_SIZE, device=DEVICE, verbose=False)
raw_metrics_v8s = results_v8s.results_dict

metrics_v8s = {
    "precision": raw_metrics_v8s.get("metrics/precision(M)", 0),
    "recall": raw_metrics_v8s.get("metrics/recall(M)", 0),
    "mAP50": raw_metrics_v8s.get("metrics/mAP50(M)", 0),
    "mAP50-95": raw_metrics_v8s.get("metrics/mAP50-95(M)", 0),
}

# --------------------------------------------
# 🔍 VALIDATE YOLOv11s-seg (Custom)
# --------------------------------------------
print("\n🔎 Validating YOLOv11s-seg (Custom)...")
model_v11s = YOLO(YOLOV11S_MODEL)
results_v11s = model_v11s.val(data=DATA_YAML, imgsz=IMG_SIZE, batch=BATCH_SIZE, device=DEVICE, verbose=False)
raw_metrics_v11s = results_v11s.results_dict

metrics_v11s = {
    "precision": raw_metrics_v11s.get("metrics/precision(M)", 0),
    "recall": raw_metrics_v11s.get("metrics/recall(M)", 0),
    "mAP50": raw_metrics_v11s.get("metrics/mAP50(M)", 0),
    "mAP50-95": raw_metrics_v11s.get("metrics/mAP50-95(M)", 0),
}

# --------------------------------------------
# 🔍 VALIDATE YOLOv5
# --------------------------------------------
print("\n🔎 Validating YOLOv5...")
val_results = yolov5_val(
    data=DATA_YAML,
    weights=YOLOV5_MODEL,
    batch_size=BATCH_SIZE,
    imgsz=IMG_SIZE,
    device=DEVICE,
    task='val',
    verbose=False
)

# YOLOv5 returns a tuple of tuples. The first tuple contains the key metrics.
val_metrics = val_results[0]

metrics_v5 = {
    "precision": float(val_metrics[0]),
    "recall": float(val_metrics[1]),
    "mAP50": float(val_metrics[2]),
    "mAP50-95": float(val_metrics[3]),
}

# --------------------------------------------
# 📊 COMPARE METRICS
# --------------------------------------------
def compute_f1(p, r):
    return 2 * (p * r) / (p + r + 1e-6)

def print_comparison(v5, v8s, v11s):
    print("\n📊 YOLOv5 vs YOLOv8s-seg vs YOLOv11s-seg Performance Comparison")
    print(f"{'Metric':<20}{'YOLOv5':<15}{'YOLOv8s-seg':<15}{'YOLOv11s-seg':<15}")
    print("=" * 65)

    # Parse each metric
    p5, r5, m5_50, m5_5095 = float(v5['precision']), float(v5['recall']), float(v5['mAP50']), float(v5['mAP50-95'])
    p8, r8, m8_50, m8_5095 = float(v8s['precision']), float(v8s['recall']), float(v8s['mAP50']), float(v8s['mAP50-95'])
    p11, r11, m11_50, m11_5095 = float(v11s['precision']), float(v11s['recall']), float(v11s['mAP50']), float(v11s['mAP50-95'])

    # Display table
    print(f"{'Precision':<20}{p5:<15.4f}{p8:<15.4f}{p11:<15.4f}")
    print(f"{'Recall':<20}{r5:<15.4f}{r8:<15.4f}{r11:<15.4f}")
    print(f"{'mAP@0.5':<20}{m5_50:<15.4f}{m8_50:<15.4f}{m11_50:<15.4f}")
    print(f"{'mAP@0.5:0.95':<20}{m5_5095:<15.4f}{m8_5095:<15.4f}{m11_5095:<15.4f}")
    print(f"{'F1 Score':<20}"
          f"{compute_f1(p5, r5):<15.4f}"
          f"{compute_f1(p8, r8):<15.4f}"
          f"{compute_f1(p11, r11):<15.4f}")

# ✅ Run the comparison
print_comparison(metrics_v5, metrics_v8s, metrics_v11s)
```

```python
import os
import time
import cv2
import torch
import numpy as np
import matplotlib.pyplot as plt
from ultralytics import YOLO
from yolov5.models.common import DetectMultiBackend
from yolov5.utils.general import non_max_suppression

# -----------------------
# 📁 Config
# -----------------------
YOLOV5_MODEL = '/kaggle/working/yolov5/runs/train/yolov5_lane_seg/weights/best.pt'
YOLOV8S_MODEL = '/kaggle/working/runs/segment/yolov8_lane_seg/weights/best.pt'
YOLOV11S_MODEL = '/kaggle/working/runs/segment/yolov11_lane_seg/weights/best.pt'

VAL_DIR = '/kaggle/input/road-lane-segmentation-imgs-and-labels/dataset/val/images'
IMG_SIZE = 640
DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')


# -----------------------
# 🔁 Benchmark Function
# -----------------------
def benchmark_model(model_name, model_type, model_path):
    image_paths = sorted([os.path.join(VAL_DIR, f) for f in os.listdir(VAL_DIR) if f.endswith(('.jpg', '.png'))])
    n_images = len(image_paths)

    if model_type == 'yolov8':
        model = YOLO(model_path)
        start = time.time()
        for img_path in image_paths:
            model.predict(source=img_path, imgsz=IMG_SIZE, device=DEVICE, verbose=False)
        total_time = time.time() - start

    elif model_type == 'yolov5':
        model = DetectMultiBackend(model_path, device=DEVICE)
        model.eval()
        model.to(DEVICE)

        start = time.time()
        for img_path in image_paths:
            img0 = cv2.imread(img_path)
            img0 = cv2.resize(img0, (IMG_SIZE, IMG_SIZE))
            img = img0[:, :, ::-1].transpose(2, 0, 1)
            img = np.ascontiguousarray(img)
            img = torch.from_numpy(img).to(DEVICE).float() / 255.0
            if img.ndim == 3:
                img = img.unsqueeze(0)

            pred = model(img, augment=False, visualize=False)
            pred = non_max_suppression(pred)
        total_time = time.time() - start

    else:
        raise ValueError("Unsupported model type")

    time_per_image = total_time / n_images
    fps = n_images / total_time

    print(f"✅ {model_name}: {time_per_image * 1000:.2f} ms/image | {fps:.2f} FPS")
    return {'name': model_name, 'inference_time': time_per_image, 'fps': fps}
```

```python
import matplotlib.pyplot as plt

# ✅ Use your actual results
results = [
    {'name': 'YOLOv5', 'inference_time': 0.01213, 'fps': 82.42},
    {'name': 'YOLOv8s-seg', 'inference_time': 0.01777, 'fps': 56.27},
    {'name': 'YOLOv11s-seg', 'inference_time': 0.01890, 'fps': 52.91}
]

def plot_inference_comparison(results, save_path=None):
    model_names = [r['name'] for r in results]
    times = [r['inference_time'] for r in results]
    fps_values = [r['fps'] for r in results]

    fig, axs = plt.subplots(1, 2, figsize=(14, 5))

    # 🔷 Inference Time
    bars1 = axs[0].bar(model_names, times, color=['#66c2a5', '#fc8d62', '#8da0cb'])
    axs[0].set_title("Average Inference Time per Image (seconds)")
    axs[0].set_ylabel("Time (s)")
    axs[0].set_xlabel("Model")
    for bar in bars1:
        height = bar.get_height()
        axs[0].text(bar.get_x() + bar.get_width() / 2, height + 0.0005, f"{height:.5f}", ha='center')

    # 🔷 FPS
    bars2 = axs[1].bar(model_names, fps_values, color=['#1b9e77', '#d95f02', '#7570b3'])
    axs[1].set_title("Frames Per Second (FPS)")
    axs[1].set_ylabel("FPS")
    axs[1].set_xlabel("Model")
    for bar in bars2:
        height = bar.get_height()
        axs[1].text(bar.get_x() + bar.get_width() / 2, height + 0.5, f"{height:.2f}", ha='center')

    plt.tight_layout()

    if save_path:
        plt.savefig(save_path, bbox_inches='tight')
    plt.show()

# ✅ Call the plotting function
plot_inference_comparison(results, save_path="inference_comparison.png")
```
